```python
def get_columns(df, group='wins', columns=['count', 'mean']):
    """
    Get a list of column names from a group in a 'GroupBy' Data Frame
    """
    return df.loc[:, group].loc[:, columns]
```

How did X card perform?

```python
def card_performance_analyzer(card_name=None, boss_name=None, df=None):
    """
    """
    if card_name == None:
        raise ValueError("Please provide a card name to analyze")
    
    if type(df) == type(None):
        pd.DataFrame(db_operations.get_table('main_table', 'all_events'))
    
    if boss_name != None:
        card_df = df.loc[df['boss']==boss_name]
    else:
        card_df = df.copy()

    
    card_df.loc[:, "cardAMT"] = card_df.deck.apply(lambda deck: helpers.card_type_in_deck(deck, [card_name]))
    return card_df.groupby('cardAMT').describe()
```

```python
tmp = card_performance_analyzer("Ghost Chase", 'Deeplands, Reguregnus', full_df)
print(get_columns(tmp, 'wins', ['count', 'mean']))
```
